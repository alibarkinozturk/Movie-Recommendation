```python
import pandas as pd
import numpy as np
```

```python
movies_df = pd.read_csv("movies.csv")
ratings_df = pd.read_csv("ratings.csv")
```

```python
movies_df
```

```python
movies_df["genres"] = movies_df["genres"].str.split("|")
df_m = movies_df.copy()
```

```python
genre = pd.DataFrame(columns = ["genre","count"])
```

```python
list_g = list()
for i in range(len(df_m)):
    for j in range(len(df_m["genres"][i])):
        list_g.append(df_m["genres"][i][j])
```

```python
g_unique = np.unique(list_g)
```

```python
for i in range(len(g_unique)):
    genre.loc[i] = [g_unique[i], int(list_g.count(g_unique[i]))]
```

```python
genre = genre[genre["genre"] != "(no genres listed)"]
genre = genre[genre["genre"] != "IMAX"]
```

```python
genre = genre.sort_values(by = "count",ascending = False).head(5)
```

```python
list_cat = genre["genre"].reset_index(drop=True)
```

```python
list_drama = list()
for i in range(len(movies_df)):
    for j in range(len(movies_df["genres"][i])):
        if(movies_df["genres"][i][j] == list_cat[0]):
            list_drama.append(movies_df["movieId"][i])
list_drama = ratings_df[ratings_df["movieId"].isin(list_drama)]

cnt = list_drama.groupby(by="movieId",as_index= False).count().sort_values(by = "userId", ascending = False)
cnt["vote_num"] = cnt["userId"]
cnt = cnt.drop(columns = "rating")

sum = list_drama.copy()
sum = sum.groupby(by= "movieId",as_index= False).sum()
sum = sum[["movieId","rating"]]

list_drama = sum.merge(cnt,how="left",on= ("movieId"))
list_drama = list_drama.drop(columns= ["timestamp","userId"])

if(list_drama["vote_num"].quantile(0.33) >=2000):
    list_drama = list_drama[list_drama["vote_num"]>list_drama["vote_num"].quantile(0.33)]

else:
    list_drama = list_drama[list_drama["vote_num"]>1999]

list_drama["recommend_rate"] =  list_drama["rating"]/list_drama["vote_num"]
list_drama = list_drama.sort_values(by = "recommend_rate", ascending = False).head(20)
l_m = list_drama["movieId"]

info = movies_df[movies_df['movieId'].isin(l_m)]
list_drama= list_drama.reset_index(drop=True)
list_drama = list_drama.merge(info, how = "left", on = "movieId")
list_drama = list_drama[["movieId","title", "recommend_rate", "vote_num", "genres"]]
```

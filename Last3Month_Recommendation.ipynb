```python
import pandas as pd
import numpy as np
import ast
import matplotlib.pyplot as plt
import re
import scipy
```

```python
movies_df = pd.read_csv("movies.csv")
ratings_df = pd.read_csv("ratings.csv")
```

```python
movies_df
```

```python
limit = (ratings_df["timestamp"].max()) - (7948800)
```

```python

#Son 3 aydaki filmleri list_m ekle
df = ratings_df[ratings_df['timestamp'] >= limit]
cnt = df.groupby(by="movieId",as_index= False).count().sort_values(by = "userId", ascending = False)
cnt["vote_num"] = cnt["userId"]
cnt = cnt.head(200)
list_m = list(cnt["movieId"])
```

```python
cnt = cnt[["movieId","vote_num"]]
```

```python
df = df[df["movieId"].isin(list_m)]
```

```python
sum = df.copy()
sum = sum.groupby(by= "movieId",as_index= False).sum()
sum = sum[["movieId","rating"]]
```

```python
df = sum.merge(cnt,how="left",on= ("movieId"))
```

```python
df["avarage_rating"] = df["rating"]/df["vote_num"]
```

```python
top12 = df.sort_values(by = "avarage_rating", ascending = False).head(20)
l_m = list(top12["movieId"])
```

```python
info = movies_df[movies_df['movieId'].isin(l_m)]
top12= top12.reset_index(drop=True)

top12 = top12.merge(info, how = "left", on = "movieId")

top12[["movieId","title", "avarage_rating", "vote_num", "genres"]]
```
